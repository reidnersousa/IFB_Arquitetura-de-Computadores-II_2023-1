```python


%%writefile letraA.cu


// verifica se ta certo

#include<stdio.h>
#include<time.h>
#include<stdlib.h>

__global__ void add(int *vetor1, int *vetor2) {
    int index = threadIdx.x + blockIdx.x * blockDim.x;
    if (vetor2[index] > 0) {
        vetor2[index] = vetor2[index] * -1;
    }
    else{
      vetor2[index] =vetor2[index]*vetor1[index];
    }
}

void random_ints(int *a, int N){
	int i;
	for (i = 0; i < N; i++){
      a[i] = (rand() % 100) - 100;

	  // a[i] = rand() % 100;
    printf("%3d, ",a[i]);
	}
  puts("");
}



#define N (1000)
#define THREADS_PER_BLOCK 8

int main(void) {
	int *vetor1, *vetor2; // host copies of a, b, c
    int *d_vetor1, *d_vetor2; // device copies of a, b, c
    int size = N * sizeof(int);
    int i;

    // Alloc space for device copies of a, b, c
    cudaMalloc((void **)&d_vetor1, size);
    cudaMalloc((void **)&d_vetor2, size);

    srand(time(NULL));

    // Alloc space for host copies of a, b, c and setup input values
    vetor1 = (int *)malloc(size);
    random_ints(vetor1, N);
    vetor2 = (int *)malloc(size);
    random_ints(vetor2, N);

    // Copy inputs to device
    cudaMemcpy(d_vetor1, vetor1, size, cudaMemcpyHostToDevice);
    cudaMemcpy(d_vetor2, vetor2, size, cudaMemcpyHostToDevice);

    // Launch add() kernel on GPU with N blocks
    add<<<((N + THREADS_PER_BLOCK - 1) / THREADS_PER_BLOCK), THREADS_PER_BLOCK>>>(d_vetor1, d_vetor2);

    // Sincronização do dispositivo
    cudaDeviceSynchronize();

    // Copy result back to host
    cudaMemcpy(vetor2, d_vetor2, size, cudaMemcpyDeviceToHost);

    for (i = 0; i < N - 1; i++) {
        printf("%-3d,", vetor2[i]);
    }
    printf("%-3d\n", vetor2[i]);

    // Cleanup
    free(vetor1);
    free(vetor2);
    cudaFree(d_vetor1);
    cudaFree(d_vetor2);

	  return 0;
}
```

Crie um programa em CUDA C que recebe um vetor contendo a quantidade (randômica) de casos diários (variando de 0 a 10.000) durante um período de 365 dias e retorne um vetor de saída contendo a média móvel desses 365 dias, levando em consideração que, para cada dia, a média dos últimos 7 dias (incluindo o dia atual). Obs: Assuma que no ano anterior não houve casos de covid (de forma que seja possível calcular a média móvel dos primeiros 6 dias do ano.



```python
%%writefile mediaMovel.cu

#include <stdio.h>
#include <stdlib.h>
#include <time.h>


// Função para calcular a média móvel em CUDA
__global__ void calcularMediaMovel(int* input, int* output, int n) {

    int idx = blockIdx.x * blockDim.x + threadIdx.x;

    if (idx < n) {
        // Variável para armazenar o somatório dos últimos 7 dias
        int soma = 0;

        // Variável para controlar o tamanho da janela deslizante
        int janela = min(7, idx + 1);

        // Calcular o somatório dos últimos 7 dias
        for (int i = 0; i < janela; i++) {
            soma += input[idx - i];
        }

        // Calcular a média móvel
        output[idx] = soma / janela;
    }


}

int main() {
    // Tamanho do vetor de entrada
    int n = 365;

     int input[N];
    int output[N];

    // Preencher o vetor de entrada com valores randômicos (entre 0 e 10000)
    srand(42);
    for (int i = 0; i < N; i++) {
        input[i] = rand() % 10001;
    }

    int* d_input;
    int* d_output;

    // Alocar memória na GPU para os vetores de entrada e saída
    cudaMalloc((void**)&d_input, N * sizeof(int));
    cudaMalloc((void**)&d_output, N * sizeof(int));

    // Copiar os dados do vetor de entrada para a GPU
    cudaMemcpy(d_input, input, N * sizeof(int), cudaMemcpyHostToDevice);

    // Chamar a função calcularMediaMovel em CUDA
    int numBlocks = (N + THREADS_PER_BLOCK - 1) / THREADS_PER_BLOCK;
    calcularMediaMovel<<<numBlocks, THREADS_PER_BLOCK>>>(d_input, d_output, N);

    // Copiar os resultados de volta para a CPU
    cudaMemcpy(output, d_output, N * sizeof(int), cudaMemcpyDeviceToHost);

    // Liberar memória da GPU
    cudaFree(d_input);
    cudaFree(d_output);

    // Imprimir o vetor de saída contendo a média móvel
    printf("Média móvel dos casos de COVID-19 por dia:\n");
    for (int i = 0; i < N; i++) {
        printf("Dia %d: %d\n", i + 1, output[i]);
    }

    return 0;

    // TODO: Preencher o vetor de entrada com valores randômicos (entre 0 e 10000)

    // TODO: Alocar memória na GPU para os vetores de entrada e saída

    // TODO: Copiar os dados do vetor de entrada para a GPU

    // TODO: Chamar a função calcularMediaMovel em CUDA

    // TODO: Copiar os resultados de volta para a CPU

    // TODO: Liberar memória da GPU

    // TODO: Imprimir o vetor de saída contendo a média móvel

    return 0;
}
```
